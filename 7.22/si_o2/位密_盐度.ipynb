```python
def dms_to_decimal(dms):
    """将度分秒转换为十进制度数"""
    dms = dms.replace('°', ' ').replace('′', ' ').replace('″', ' ').replace('"E', ' ').replace('"N', ' ').replace('"S', ' ')
    parts = dms.split()
    degrees = float(parts[0])
    minutes = float(parts[1]) if len(parts) > 1 else 0
    seconds = float(parts[2]) if len(parts) > 2 else 0
    decimal = degrees + (minutes / 60) + (seconds / 3600)
    return decimal
```

```python
def draw(path_names):
    plt.rcParams['font.size'] = 18
    all_colors = ['#8B8970', '#FFA54F', '#3CB371', '#8B2500', '#E066FF']
    all_colors.reverse()
    all_markers = ['o', 'D', 'X', '*', '^']
    all_markers.reverse()
    colors = all_colors[:len(path_names)]
    
    fig = plt.figure(figsize=(10, 10))
    ax1 = fig.add_subplot(111)
    
    for idx, path_name in enumerate(path_names):
        hai_qu_wenjianjia = os.listdir(rf'D:\ze_1112\最新数据\分海区\{path_name}')
        first_file = True
        for i in hai_qu_wenjianjia:
            if i.endswith('.xls'):
                df = pd.read_excel(os.path.join(rf'D:\ze_1112\最新数据\分海区\{path_name}', i), skiprows=3, usecols=[1,5,6,8,10,11,13,16], names=['站位名', '纬度', '经度', '深度', '压力', '水温', '盐度', '溶解氧'])
                df_1 = df[df['深度'] >= 2000]
                depth = df['深度']
                depth_1 = df_1['深度'].values.astype(float)
                o2 = df['溶解氧']
                o2_1 = df_1['溶解氧'].values.astype(float)
                salinity = df['盐度']
                salinity_1 = df_1['盐度'].values.astype(float)
                t = df['水温']
                t_1 = df_1['水温'].values.astype(float)
                p = df['压力'] * 0.1 - 10.1325
                p_1 = df_1['压力'].values * 0.1 - 10.1325
                lon = df['经度'].apply(dms_to_decimal)
                lon_1 = df_1['经度'].apply(dms_to_decimal).values.astype(float)
                lat = df['纬度'].apply(dms_to_decimal)
                lat_1 = df_1['纬度'].apply(dms_to_decimal).values.astype(float)
                
                ab_salinity = gsw.SA_from_SP(salinity_1, p_1, lon_1, lat_1)
                pt = gsw.pt0_from_t(ab_salinity, t_1, p_1)
                CT = gsw.CT_from_pt(ab_salinity, pt)
                sigma = gsw.sigma0(ab_salinity, CT)
                if first_file:
                    ax1.scatter(sigma, o2_1, s=8, marker=all_markers[idx], c=colors[idx], alpha=0.5, label=path_name)
                    first_file = False
                else:
                    ax1.scatter(sigma, o2_1, s=8, marker=all_markers[idx], c=colors[idx], alpha=0.5)
        
    # ax1.set_ylim(1, 2.5)
    ax1.set_xlabel('σ0')
    ax1.set_ylabel('o2 (μmol/kg)')
    ax1.spines[:].set_linewidth(2)
    
    # # 绘制位势密度等值线
    # salinity_range = np.linspace(34.6, 34.7, 10)
    # temperature_range = np.linspace(1, 2.5, 10)
    # salinity_grid, temperature_grid = np.meshgrid(salinity_range, temperature_range)
    # density_grid = gsw.sigma0(salinity_grid, temperature_grid)

    # cs = ax1.contour(salinity_grid, temperature_grid, density_grid, levels=np.arange(27.5, 27.7, 0.01), colors='gray', linewidths=0.5, linestyles='dashed')
    # ax1.clabel(cs, fmt='%2.2f', colors='black', fontsize=8)
    
    plt.legend(markerscale=2)
    plt.savefig('sigma_o2.png', bbox_inches='tight', dpi=300)
    plt.close()
    # plt.show()
    
draw(wen_jian_jia)
```
