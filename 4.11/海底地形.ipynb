```python
elevation = data['deptho']

# 设置地图投影
projection = ccrs.PlateCarree()

# 创建图形和轴对象
fig, ax = plt.subplots(figsize=(12, 8), subplot_kw={'projection': projection})

# 添加海底地形图
contourf = ax.contourf(elevation.longitude, elevation.latitude, elevation, transform=ccrs.PlateCarree(), cmap='ocean_r')

# 创建自定义色标
cbar_ax = fig.add_axes([0.15, 0.1, 0.7, 0.05])  # 调整色标的位置和大小
norm = plt.Normalize(vmin=elevation.min(), vmax=elevation.max())
cb = colorbar.ColorbarBase(cbar_ax, cmap='ocean_r', norm=norm, orientation='horizontal')
cb.set_label('Depth (m)')
cb.set_ticks(np.linspace(0, 10000, num=11))  # 设置色标的刻度
# cb.set_ticklabels(np.linspace(elevation.max(), elevation.min(), num=5))  # 反转刻度标签

# 设置图标题
ax.set_title('Sea Floor Topography')

# 添加海岸线
ax.coastlines()

# 添加国界线
ax.add_feature(cfeature.BORDERS)

# 显示图形
plt.show()

```
